# 2. 특정 시술 유형 세분화 

```python
procedures = [
    "IVF", "ICSI", "IUI", "ICI", "GIFT", "FER", "Generic DI", "IVI",
    "BLASTOCYST", "AH", "Unknown"
]

def process_procedure(procedure):
    if pd.isna(procedure):  # If the value is NaN, return NaN for all columns
        return {p: np.nan for p in procedures} | {'Contains_/': np.nan, 'Contains_:': np.nan}

    procedure_set = set()
    
    contains_slash = '/' in procedure  # Checks if '/' is present
    contains_colon = ':' in procedure  # Checks if ':' is present

    # Split by '/' (multiple methods)
    for part in procedure.split('/'):
        # Split by ':' (sub-methods)
        for sub_part in part.split(':'):
            procedure_set.add(sub_part.strip())  # Add cleaned method
    
    # Create a dictionary for one-hot encoding and indicators for `/` and `:`
    return {p: (1 if p in procedure_set else 0) for p in procedures} | {'Contains_/': int(contains_slash), 'Contains_:': int(contains_colon)}


# Apply transformation to each row
encoded_df = df['특정 시술 유형'].apply(lambda x: pd.Series(process_procedure(x)))

df = pd.concat([df.drop('특정 시술 유형', axis = 1), encoded_df], axis = 1)
```

```python
def missing(df_drop):
    # Fill missing values in binary columns with the mode (most frequent value)
    for col in df_drop.select_dtypes(include=['number']):  # Select numeric columns
        if set(df_drop[col].dropna().unique()).issubset({0, 1}):  # Check if it's binary
            df_drop[col].fillna(df_drop[col].mode()[0], inplace=True)  # Replace NaN with mode

    # Fill missing values in categorical columns with the mode
    for col in df_drop.select_dtypes(include=['object']):  
        df_drop[col].fillna(df_drop[col].mode()[0], inplace=True)

    # Fill missing values in numeric columns with the mean
    df_drop.fillna(df_drop.mean(numeric_only=True), inplace=True)
    return df_drop
```
